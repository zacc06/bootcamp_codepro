```python
#Ejemplo del juego
def obtener_fila_verifficada(palabra_a_encontrar, palabra_ingresada):
    cant_letras=5
    
    #Creamos una lista vacia para almacenar el resultado
    letras_verificadas=[]
    
    for i in range (cant_letras):
        comparacion= palabra_a_encontrar[i]== palabra_ingresada[i]
        ubi_letra=palabra_ingresada[i] in palabra_a_encontrar
        
        if comparacion:
            letras_verificadas.append("["+palabra_ingresada[i]+"]")
            
        elif ubi_letra:
            letras_verificadas.append("("+palabra_ingresada[i]+")")
        else:
            letras_verificadas.append(palabra_ingresada[i])
    return letras_verificadas
    
resultado=obtener_fila_verifficada("perro","juego")
print(resultado)
```
