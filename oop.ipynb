```python
#Ejercicio1
class Animales:
    especie="Aves"
    especie="Peces"
    #Atributo de clase
    def __init__(self,nombre,color):
        self.nombre=nombre
        self.color=color
        
#Instancia de clase        
ave1=Animales("Papu","verde")
print("Me llamo ",ave1.nombre," y soy un ave de color ",ave1.color)
pescado=Animales("piraña","negro")
print("Me llamo ",pescado.nombre," y soy un pescado de color ",pescado.color)
```

```python
#Ejercicio2 
class Animales:
    especie = "murcielagos"   # Atributo de clase

    def __init__(self,nombre,color):
        self.nombre = nombre
        self.color = color
        
    def sleep(self):
        print("Mientras duerme")
        
    def colgarse(self):
        print("Esta colgado")
        
mamifero1 = Animales("Murcielago","negro")

print("El", mamifero1.nombre, "de color", mamifero1.color)

mamifero1.sleep()
mamifero1.colgarse()
```

```python
class Madre:
    especie="humano"
    
    def __init__(self,caminar,correr):
        self.caminar=caminar
        self.correr=correr
        
class Hijos(Madre):
    
    def __init__(self,gatear,llorar):
        self.gatear=gatear
        self.llorar=llorar

acciones=Hijos("Gatear","Llorar")

print(acciones.gatear)
print(acciones.llorar)
```
